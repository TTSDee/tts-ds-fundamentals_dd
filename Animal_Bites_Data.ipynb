```python
import pandas as pd
```

```python
#importing raw data downloaded from Kaggle#
df= pd.read_csv('Health_AnimalBites_2.csv')
print(df)
```

```python
#returning data 
df.head(205)
```

```python
#Delete rows of
df = df.drop(labels=["color", "vaccination_yrs", "vaccination_date", "AdvIssuedYNDesc", "quarantine_date", "DispositionIDDesc","head_sent_date", "release_date", "ResultsIDDesc"], axis=1) 
df.head(200)
```

```python
#How many missing values do we have#
df.isnull().sum()
```

```python
#Delete the rows that contain missing values#

df = df.dropna(axis = 0, how = "any")
df.head(210)
```

```python
#Resetting the index, so it starts with 0 again#
df = df.reset_index(drop = True)
df.head(210)
```

```python
df.to_csv("Health_AnimalBites_2.csv") 
```
